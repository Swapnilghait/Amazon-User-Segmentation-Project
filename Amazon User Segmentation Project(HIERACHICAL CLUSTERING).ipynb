# AMAZON HOME PAGE PROJECT (HIERACHICAL CLUSTERING)

**IMPORTING LIBRARIES AND DATASET**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("/content/drive/MyDrive/PREPINSTA /MACHINE LEARNING /5.CLUSTERING /DATASET /Amazon.com Clusturing Model (Pr (1).csv")
data_set
```

```python
# taking the rows from dataset for maing the clustors and visualizing
# we will select the income and rating for the visualization

X = data_set.iloc[:, [2, 4]].values
print(X)
```

**FINDING OPTIMAL NO OF CLUSTERS USING THE DENDOGRAM**

```python
# we will import from the scipy library
import scipy.cluster.hierarchy as sch
# method is used is "ward"
dendrogram = sch.dendrogram(sch.linkage (X , method = "ward"))

plt.title("DENDOGRAM")
plt.xlabel("AGE")
plt.ylabel("salary")
plt.show()
```

**Hierarchical clustering model training on training data**

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 6, affinity = 'euclidean', linkage = 'ward')
yhc = hc.fit_predict(X)
```

**VISUALIZING THE CLUSTER**

```python
plt.scatter(X[yhc == 0, 0], X[yhc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[yhc == 1, 0], X[yhc == 1, 1], s = 100, c = 'cyan', label = 'Cluster 2')
plt.scatter(X[yhc == 2, 0], X[yhc == 2, 1], s = 100, c = 'blue', label = 'Cluster 3')
plt.scatter(X[yhc == 3, 0], X[yhc == 3, 1], s = 100, c = 'magenta', label = 'Cluster 4')
plt.scatter(X[yhc == 4, 0], X[yhc == 4, 1], s = 100, c = 'yellow', label = 'Cluster 5')
plt.scatter(X[yhc == 5, 0], X[yhc == 5, 1], s = 100, c = 'black', label = 'Cluster 6')

plt.title('Amazon Users')
plt.xlabel('Age')
plt.ylabel('Rating')

plt.legend()
plt.show()
```
